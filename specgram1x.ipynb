```python
import numpy as np
import math
from array import *
from scipy.io import loadmat
from numpy.polynomial import Polynomial
import matplotlib.pyplot as plt


def plotting(freq, c_gr, c_gr_mod):
    freq = np.delete(freq, len(freq) - 1)
    plt.plot(freq, c_gr[:999], color='g', label='')
    plt.plot(freq, c_gr_mod, color='r', label='')
    plt.xlabel(", ")
    plt.ylabel(" , /")
    plt.title("       ")
    plt.legend()
    plt.grid()
    plt.show()
    return 0

def c_phase_group_new(h, ro, E, nu):
    output = []
    freq = freq_n
    for i in range(len(freq)):
        values = []
        w = 2 * np.pi * freq[i]
        a5 = 1
        a4 = h * w * ro / 1000 - 9.8 / w
        a3 = 0
        a2 = 0 
        a1 = -E * (w ** 3) * (h ** 3) / (12000 * (1 - nu ** 2))

        p = [a5, a4, a3, a2, a1]
        roots = np.roots(p)
        real = np.isreal(roots)
        for i in range(len(roots)):
            if real[i] == True and roots[i] > 0:
                values.append(i)
        roots = abs(roots[values])[0]
        output.append(roots)
    print(output)
    roots_gr = []

    for i in range(len(freq) - 1):
        c_group = ((1/output[i]) + ((2*np.pi*freq[i]) / ((output[i])**2)) * (output[i+1] - output[i])/(freq[i+1] - freq[i]))**(-1)        
        #c_group = output[i] / (1 + (2 * np.pi * freq[i] / output[i]) * (output[i + 1] - output[i]) / (2 * np.pi * freq[2] - 2 * np.pi * freq[1]))
        roots_gr.append(c_group)
    return roots_gr
```

```python
c_gr_model = c_phase_group_new(0.5, 900, 4.5*10**9, 0.33)

plt.scatter(x=freq_n[:30], y=c_gr_model[:30])
```
